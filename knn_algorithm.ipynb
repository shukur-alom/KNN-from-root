```python
import pandas as pd # pandas==1.1.4
import numpy as np # numpy==1.19.3
from sklearn.model_selection import train_test_split #scikit-learn==0.22.2
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/train.csv")
```

```python
x = data.drop("price_range",axis=1).to_numpy()
y = data["price_range"].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
def knn(new_data,k,x,y):
  new_data = np.array(new_data)
  distance = []
  result = []
  for i in x:
    distance.append(np.linalg.norm(i - new_data))
  for i in range(k):
    index = distance.index(min(distance))
    result.append(y[index])
    distance[index] = max(distance)

  return max(result)
```

```python
# predict
new_data = list(x[10])
print(knn(new_data,k=15,x=X_train,y=y_train))
```

```python
y[10]
```

```python
y_ped = []
for i in range(len(X_test)):
  new_data = list(X_test[i])
  y_ped.append(knn(new_data,k=15,x=X_train,y=y_train))
  print(f"Predict : {knn(new_data,k=15,x=X_train,y=y_train)}.. Actual data : {y_test[i]}")
```

```python
print(f"Accuracy is : {accuracy_score(y_test, y_ped)*100}%")
```
